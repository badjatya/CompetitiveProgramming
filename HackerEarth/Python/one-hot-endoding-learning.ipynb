```python
import pandas as pd 
import numpy as np 
import matplotlib as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
dummiesvariable = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df,dummiesvariable], axis='columns')
merged
```

```python
final = merged.drop(['town','west windsor'], axis="columns")
final
```

```python
reg = linear_model.LinearRegression()
dataset = final.drop('price', axis='columns')
dataset
```

```python
y=final.price
```

```python
reg.fit(dataset, y)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.score(dataset,y)
```
