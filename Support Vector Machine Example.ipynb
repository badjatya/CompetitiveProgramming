```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
dir(data)
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
df['Target'] = data.target
df.head()
```

```python
df[df.Target == 1].count()
```

```python
df['Flower_name'] = df.Target.apply(lambda x: data.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.Target==0]
df1 = df[df.Target==1]
df2 = df[df.Target==2]
```

```python
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], color='red' ,marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], color='blue' ,marker='+')
```

```python
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color='red' ,marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color='blue' ,marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
target = df.Target
inputdata = df.drop(['Target','Flower_name'], axis='columns')
```

```python
x_train, x_test, y_train, y_test=train_test_split(inputdata,target, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
